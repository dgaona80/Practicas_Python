```python
import pandas as pd

df = pd.read_csv("clientes.csv")

#Nan = valor nulo o faltante
print("--------------------------------Lectura del CSV--------------------------------")
print(df.head(),"\n")
print("\n--------------------------------resumen del df--------------------------------")
print("\n",df.info())
print("\n--------------------------------detalle de nulos--------------------------------")
print(df.isnull()) #calcular total de campos nulos
print("\n--------------------------------resumen de nulos--------------------------------")
print(df.isnull().sum()) #calcular total de campos nulos

print("-------------------------------- Valores nulos eliminados: --------------------------------")

df_limpio = df.dropna()
print(df_limpio)
print("\n--------------------------------Rellenar los datos nulos --------------------------------\n")

#rellenar la edad y gasto con la media
"""
media_edad = df["Edad"].mean()
df["Edad"] = df["Edad"].fillna(media_edad) #llena los datos faltantes con la media de edad
print("Resultado del DF con la edad calculada con la media (si hay nulos)\n",df,"\n")
"""

media_gasto = df["Gasto"].mean()
df["Gasto"] = df["Gasto"].fillna(media_edad) #llena los datos faltantes con la media del casto
print("Resultado del DF con el Gasto calculado con la media (si hay nulos)\n",df)

#uso de la moda para rellenar valores
modo_ciudad = df["Ciudad"].mode()[0] #el que más se repite, sólo trae 1
df["Ciudad"] = df["Ciudad"].fillna(modo_ciudad)
print("\nResultado del DF rellenado con la ciudad calculada (MODA))\n",df)

#uso de la moda para rellenar valores
modo_ciudad = df["Cliente"].mode()[0] #el que más se repite, sólo trae 1
df["Cliente"] = df["Cliente"].fillna(modo_ciudad)
print("\nResultado del DF rellenado con la Cliente calculado (MODA))\n",df)

print("\nResultado de datos NULOS RELLENOS por REGRESIÓN SIMPLE:")
df["Edad"] = df.groupby('Ciudad')['Edad'].transform(
    lambda x: x.fillna(x.median())
)

print(df)
print("-------------------------------- --------------------------------")
print("-------------------------------- --------------------------------")
```
