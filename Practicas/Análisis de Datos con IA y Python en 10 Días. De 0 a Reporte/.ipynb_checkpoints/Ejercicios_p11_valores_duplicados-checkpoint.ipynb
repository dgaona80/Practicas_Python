```python
import pandas as pd

df = pd.read_csv("ventas/clientes.csv", encoding = "latin-1")

print("-----------------------df original:------------------------\n",df.head(),"\n")

print("-----------------------df original carcterísticas:------------------------\n",df.info(),"\n")

#Normalizar ciudad

for col in df.select_dtypes(include='object'):
    df[col] = df[col].str.lower().str.strip()
    df[col].info()

print("-----------------------df normalizando a minúsculas:-----------------------\n",df.head(10),"\n")


print("ordernar duplicados\n")

duplicados_det = df.duplicated(subset=["Nombre","Correo","Telefono","Ciudad"], keep=False) #no importa el orden

duplicados_sum = df.duplicated(subset=["Nombre","Correo","Telefono","Ciudad"], keep=False).sum() #no importa el orden
print(duplicados_det,"\nTotal de duplicados:",duplicados_det.sum())

df = df.drop_duplicates(subset=["Nombre","Correo","Telefono","Ciudad"])

print("-----------------------Eliminación de duplicados:-----------------------\n",df,"\n")
```
