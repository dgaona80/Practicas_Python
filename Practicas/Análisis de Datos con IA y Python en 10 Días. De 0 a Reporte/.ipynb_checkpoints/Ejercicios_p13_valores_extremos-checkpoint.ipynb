```python
import pandas as pd
import numpy as np

df = pd.read_csv("ventas/productos_2.csv", encoding = "Latin-1")

print("--------------Datos originales---------------------\n",df,"\n")

print("--------------Datos descriptivos---------------------\n",df["Precio"].describe(),"\n")

print("-------------- Detección con IQR---------------------\n")

Q1 = df["Precio"].quantile(0.25)
Q3 = df["Precio"].quantile(0.75)

IQR = Q3 - Q1

print(f'IQR: {IQR}\n')

limite_inferior = Q1 -1.5 * IQR
limite_superior = Q3 +1.5 * IQR

print("-------------- Cálculo de límites ---------------------\n")
print(f'Lím. Inferior: {limite_inferior}, lím. Superior: {limite_superior}')

#print("-------------- Cálculo de Outliers ---------------------\n")

outliers_iqr = df[(df["Precio"] < limite_inferior) | (df["Precio"]> limite_superior)]

print(f'-----------------Outliers (valores fuera de los límites:--------------- \n{outliers_iqr}')

#filtro de Z-SCORE

print(f'\n----------------- filtro de Z-SCORE: -----------------')

media = df["Precio"].mean()
std = df["Precio"].std()
df["Z-SCORE"] = (df["Precio"] - media)/ std

outliers_z = df[np.abs(df["Z-SCORE"])>3]
print(f'\n {df[["Precio","Z-SCORE"]]}')

print("\n",df)

print(f'\n----------------- Datos con valores OUTLIERS ELIMINADOS: -----------------')

df_limpio = df[(df["Precio"]>=limite_inferior) & (df["Precio"] <= limite_superior)]

print(f'\n {df_limpio}')

print(f'\n----------------- Datos con valores OUTLIERS CORREGIDOS: -----------------')

#df.loc[df["Precio"] == 5000, "Precio"] = 50.0
#df.loc[df["Precio"] == -10, "Precio"] = 10.0

print(f'\n {df}')


print(f'\n----------------- Datos con valores OUTLIERS ETIQUEDADOS: -----------------')

df["Es_Outlier"] = ((df["Precio"] <limite_inferior) | (df["Precio"]> limite_superior))

print(f'\n {df}')
```
