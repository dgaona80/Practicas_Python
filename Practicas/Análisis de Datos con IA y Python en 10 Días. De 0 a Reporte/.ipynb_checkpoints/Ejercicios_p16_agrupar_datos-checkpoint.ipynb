```python
import pandas as pd

data = {
    "Estudiante" : ["Ana","Luis","Pedro","Laura","Carlos","Julia","Andrés","Marta","Juan","Daniel"],
    "Curso" : ["Python","Excel","Python","Excel","Python","SQL","SQL","Excel","Python","Python"],
    "Región" : ["Latam","España","Latam","España","Latam","España","Latam","España","Latam","España"],
    "Calificación" : [85,90,78,95,88,70,75,92,89,83],
    "Horas_Estudio" : [10,8,12,7,9,5,6,8,11,7]
}

df = pd.DataFrame(data)

print("------------- Dataframe original:----------",df)

calif_x_curso = df.groupby("Curso")["Calificación"].mean()

print("\n------------- Promedio de calificación por curso:----------",calif_x_curso)


calif_CursoEstd = df.groupby(["Curso","Estudiante"])["Calificación"].mean()


print("\n------------- Promedio de calificación por curso x estudiante----------",calif_CursoEstd)

Indicadores_curso = df.groupby(["Curso","Estudiante"])["Calificación"].agg(["mean","max","min","count"])

print("\n------------- Tabla de indicadores x curso ----------\n",Indicadores_curso)


print("\n------------- Describe ----------\n",df.describe())


Indicadores_horasestudio = df.groupby(["Curso","Estudiante"])[["Calificación","Horas_Estudio"]].agg({
    "Calificación" : "mean",
    "Horas_Estudio" : "sum"})


print("\n------------- Tabla de indicadores horas de estudio ----------\n",Indicadores_horasestudio)


```
