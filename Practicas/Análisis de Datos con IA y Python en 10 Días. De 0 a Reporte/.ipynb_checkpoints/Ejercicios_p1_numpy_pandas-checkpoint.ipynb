```python
"""
    LECCIÓN 3 CURSO ANALISIS DE DATOS CON IA EN 10 DIAS
    
    - Sumar incentivos de + 5 unidades vendidas pr una campaña
    - Calcular el doble de lo que habríasmo vendidos si hubieramos tenido el stock completo
    - Saber el promedio de ventas
    - Identificar el día con más y menos ventas
    - Ver los totales por semana
    
"""
# Ejercios de Numpy
  
import numpy as np
import pandas as pd

# Datos reales de los días

ventas = np.array([12,15,13,10,18,20,22,17,16,19])

print("Ventas totles (dia 1 al 10): \n",ventas,'\n')


ventas_con_incentivo = ventas + 5
print("Ventas con campaña +5 de incentivo: \n",ventas_con_incentivo,'\n')

ventas_dobladas = ventas * 2
print("Ventas potenciales con más stock (dobladas): \n",ventas_dobladas,'\n')

#promedio de ventas

promedio_ventas = np.mean(ventas)

print("Promedio de ventas:$", promedio_ventas,'\n')

#Máximos y mínimos

dia_max = np.max(ventas)
dia_min = np.min(ventas)

print(f'Máximo de ventas en un día:${dia_max}')
print(f'Máximo de ventas en un día:${dia_min}','\n')

ventas_semana = ventas.reshape(2,5)
print(f'Ventas en semanas (2 semanas x 5 días: \n {ventas_semana}')
print('')
print('')
print('')


#Pandas

df = pd.read_csv('orders.csv')
print("Primeros registros de datos:")
#print primeros 5 registros
print(df.head(5),'\n')

#información del archivo
print("Información del archivo: \n")
print(df.info())

#estadísticas iniciales del archivo
print("Estadísticas del archivo CSV: \n")
print(df.describe())

#Colmunas disponibles
print("columnas disponibles: \n")
print(df.columns)

#Valores unicos
print("Valores únicos del archivo:\n")
print(df['category product'].unique())
print("\n")
print("\n")
print("\n")
print("\n")
print("\n")
print("\n")
print("\n")
```
