```python
import pandas as pd
import numpy as np
from scipy import stats

data = {
    "Mes" : ["Enero","Febrero","Marzo","Abril","Mayo","Junio","Julio","Agosto","Septiembre","Octubre","Noviembre","Diciembre"],
    "Ventas" : [1200,1500,1300,1700,2000,1800,2200,2100,1900,2500,2400,2300]   
}

df = pd.DataFrame(data)


print("------------- DATAFRAME ORIGINAL:----------\n",df,"\n\n",df.describe())

#Medidas de tendencias
media = np.mean(df["Ventas"])
mediana = np.median(df["Ventas"])
moda1 = df["Ventas"].mode()
moda2 = stats.mode(df["Ventas"], keepdims=True).mode[0]
moda3 = df["Ventas"].mode()[0]


#medidas de dispersión
varianza = np.var(df["Ventas"],ddof=1)
desviacion_std = np.std(df["Ventas"],ddof=1)
rango =np.max(df["Ventas"]) - np.min(df["Ventas"])

print("------------- CÁLCULOS DE MEDIDAS DE TENDENCIA CENTRAL----------\n")

print(f'Media: {media}')
print(f'Mediana: {mediana}')
print(f'Moda 1(mode):\n {moda1}')
print(f'Moda 2(stats): {moda2}\n')
print(f'Moda 3(mode -limpia): {moda2}\n')

print("------------- CÁLCULOS DE MEDIDAS DE DISPERSIÓN----------\n")

print(f'Varianza: {varianza}')
print(f'Desviación estándar: {desviacion_std}')
print(f'Rango: {rango}')

#print(f': {}')

print('-----------------------CREAR TABLA RESUMEN------------------')
resumen = pd.DataFrame({
    "Métrica" : ["Media","Mediana","Moda","Varianza","Desv. Estándar","Rango"],
    "Valor" : [media,mediana,moda3,varianza,desviacion_std,rango]
})

print("\n Resunen de ventas anuales:")
print(resumen.to_string(index=False))

#interpretación ejecutiva

print("\n\n〽️Interpretación Ejecutiva:")
print(f'◽ En promedio, las ventas fueron de: ${media:,.2f} mensuales.')
print(f'◽ El mes con mayor ventas fue: {df.loc[df["Ventas"].idxmax(),'Mes']} con ${df["Ventas"].max():,.2f}')
print(f'◽ El mes con menor ventas fue: {df.loc[df["Ventas"].idxmin(),'Mes']} con ${df["Ventas"].min():,.2f}')
print(f'◽ La desviación estándar de ${desviacion_std:,.2f}') # indica una variación {'alta if }')
print(f'◽')
print(f'◽')


```
