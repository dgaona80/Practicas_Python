```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


#Datos de ejemplo: edades (distribución) y ventas por producto(proporciones)

edades = [22,25,47,52,46,35,27,30,40,45,50,60,65,70,75,80,85,90]

df = pd.DataFrame({'Edad':edades})

ventas ={
    'Producto' : ['Camisas','Pantalones','Zapatos','Accesorios'],
    'Cantidad': [150,200,120,80]}

dfv = pd.DataFrame(ventas)

#Histograma

plt.figure(figsize=(7,4))
sns.histplot(
    df['Edad'],
    bins = 8,
    color = 'skyblue',
    edgecolor='black'
)
plt.title('histograma de distribución de edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

#Gráfica de pay

plt.figure(figsize=(6,6))
plt.pie(dfv['Cantidad'],
       labels=dfv['Producto'],
       autopct='%1.1f%%',
       startangle=90,
       colors=sns.color_palette('pastel'))
plt.title('% de ventas por producto')
plt.show()

    
```
