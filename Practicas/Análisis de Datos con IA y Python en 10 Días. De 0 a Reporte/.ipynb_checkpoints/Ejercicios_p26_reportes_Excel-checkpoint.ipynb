```python
import pandas as pd
import xlsxwriter 

df = pd.DataFrame({
    'Producto' : ['A','B','C','D'],
    'Ventas' : [12500,2300,17500,19500]
})


writer = pd.ExcelWriter('Reporte_ventas_xls.xlsx', engine='xlsxwriter')
df.to_excel(writer,sheet_name='Resumen', index=False)

workbook = writer.book
worksheet = writer.sheets['Resumen']

#Formato para toda la página o libro
header_fat = workbook.add_format({'bold': True, 'font_color':'white', 'bg_color':'#4F81BD'}) #agregarle formato
num_fmt = workbook.add_format({'num_format':'#,##0'})

#formato para encabezados

for col_num, value in enumerate(df.columns.values):
    worksheet.write(0,col_num,value, header_fat)

worksheet.set_column(0,0,15)
worksheet.set_column(1,1,18,num_fmt)

#sacar totales

total_row =len(df) + 1

worksheet.write(total_row,0,'Total')
worksheet.write_formula(total_row,1,f'=SUM(B2:B{total_row})',num_fmt)

chart = workbook.add_chart({'type':'column'})
chart.add_series({
    'name': 'ventas por producto',
    'categories' : ['Resumen',1,0,len(df),0],
    'values' : ['Resumen',1,1,len(df),1],
    'data_labels' : {'value': True, 'num_format': '#,##0'}
})

chart.set_title({'name': 'ventas por producto'})
chart.set_x_axis({'name':'Producto'})
chart.set_y_axis({'name':'Ventas', 'num_format':'#,##0'})
worksheet.insert_chart('D2',chart)
writer.close()


#C:\Users\USER\Documents\CURSO_PYTHON\Análisis de Datos con IA y Python en 10 Días. De 0 a Reporte


```
