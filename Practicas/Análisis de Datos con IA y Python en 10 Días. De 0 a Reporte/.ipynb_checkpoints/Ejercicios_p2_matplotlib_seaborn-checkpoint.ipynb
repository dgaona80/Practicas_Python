```python
"""

  DIA 2 EJERCICIO 2: USO DE MATPLOTLIB Y SEABORN

"""

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Pandas

df = pd.read_csv('orders.csv')

#sustituir NaN por 1 en qty
df['qty'].fillna(1, inplace=True)

print("Primeros registros de datos:")
#print primeros 5 registros
print(df.head(5),'\n')

#información del archivo
print("Información del archivo: \n")
print(df.info())

#Ventas por producto

#Establecemos el tamaño de la grafica
plt.figure(figsize=(10,5))

#Grafico de barras
                #EJE X                  EJE Y            COLOR DE BARRA
plt.bar(df['category product'], df['unit_price'], color = 'skyblue') #dibujamos una barra por cada producto (eje Y=unit price, eje X=category product)

# Rotar etiquetas del eje X
plt.xticks(rotation=45, ha='right')  # ha='right' alinea a la derecha

# Ajustar márgenes para que no se corten los textos
plt.tight_layout()

#Esto hay que practicarlo
# Formatear etiquetas largas en varias líneas
#categorias_formateadas = [textwrap.fill(cat, width=15) for cat in categorias]
# Crear gráfica de barras
#plt.bar(categorias_formateadas, valores)
# Rotar etiquetas si quieres mayor legibilidad
#plt.xticks(rotation=0, ha='center')  # puedes probar con rotation=30 o 45

#Agregar títulos y etiquetas
plt.title('VENTAS POR CATEGORÍA DEL PRODUCTO')
plt.xlabel('CAT. PRODUCTO')
plt.ylabel('UNIT PRICE')


#mostrar grafica
plt.show()

productos_unicos = df['product_name'].unique()
print(f'Productos unicos para gráfica: \n {productos_unicos}')

### GRAFICA 2
#Distribución de productos histograma
plt.figure(figsize=(10,5))
plt.hist(df['unit_price'], bins=5, color='lightgreen', edgecolor='black')
#Agregar títulos y etiquetas
plt.title('VENTAS POR NOMBRE DEL PRODUCTO')
plt.xlabel('PRODUCTO')
plt.ylabel('UNIT PRICE')


#mostrar grafica
plt.show()


#Crear Gráficas en Seaborn barras

plt.figure(figsize=(15,5))

sns.barplot(x = 'product_name', y = 'unit_price', data=df, palette='viridis')
# Rotar etiquetas del eje X
plt.xticks(rotation=45, ha='right')  # ha='right' alinea a la derecha

plt.title('CANTIDADES POR PRODUCTO (SEABORN)')
plt.xlabel('Productos')
plt.ylabel('Pedidos')

plt.show()


# Gráfico 2 Seaborns

plt.figure(figsize=(15,5))

#sns.barplot(x = 'qty', y = 'unit_price', data=df, hue = 'customer_city', style = 'customer_city', s=100)
sns.barplot(x='qty', y='unit_price', data=df, hue='customer_city')

# Rotar etiquetas del eje X
plt.xticks(rotation=45, ha='right')

plt.title('PEDIDOS POR PRODUCTO (SEABORN)')
plt.legend(title = 'Ciudad')
plt.xlabel('Cantidad')
plt.ylabel('Precio')

plt.show()



# Gráfico 3 Seaborns

plt.figure(figsize=(15,5))

sns.scatterplot(x = 'qty', y = 'unit_price', data=df, hue = 'customer_city', style = 'customer_city', s=100)

# Rotar etiquetas del eje X
plt.xticks(rotation=45, ha='right')  # ha='right' alinea a la derecha

plt.title('PEDIDOS POR PRODUCTO (SEABORN)')
plt.legend(title = 'Ciudad')
plt.xlabel('Cantidad')
plt.ylabel('Precio')

plt.show()

```
