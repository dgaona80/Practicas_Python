```python
import pandas as pd
from pandas import json_normalize

import json
with open("ventas/ventas_consolidadas.json","r", encoding="utf-8" ) as archivo_crudo:
    datos_estructura = archivo_crudo.read() # leer estructura del JSON
     
print("JSON LEIDO\n Datos:\n")
#print(datos_estructura)



with open("ventas/ventas_consolidadas.json","r", encoding="utf-8" ) as archivo_normalizado:
    datos =   json.load(archivo_normalizado) #Leer los datos
    

# --- Normalizar cada sección ---
# Metadata → un solo registro
df_metadata = pd.DataFrame([datos["metadata"]])

# Resumen → un solo registro
df_resumen = pd.DataFrame([datos["resumen"]])

df_ventas = json_normalize(datos,record_path="ventas")

ruta_xls = "ventas/datos_JSON.xlsx"

#genera una pestaña por cada sección
with pd.ExcelWriter(ruta_xls, engine="openpyxl") as writer:
    df_metadata.to_excel(writer, sheet_name="Metadata", index=False)
    df_resumen.to_excel(writer, sheet_name="Resumen", index=False)
    df_ventas.to_excel(writer, sheet_name="Ventas", index=False)

print(f'\n Datos Normalizados Legibles enviados a Excel:\n')


```
