```python
import requests
import pandas as pd

#API publica

url = "https://restcountries.com/v3.1/all?fields=name,region,area,population"

headers = {
    "User-Agent": "Mozilla/5.0"

}

try:
    response = requests.get(url,headers=headers)
    response.raise_for_status()
    print("Conexion exitosa con la API")
    
    datos = response.json()
    
    paises = []
    for pais in datos:
        paises.append({
            "Nombre": pais.get("name",{}).get("common",None),  #si no encuentra nada en name, se va a common
            "Region": pais.get("region",None),
            "Area (KM2)" : pais.get("area",None),
            "Poblacion" : pais.get("population",None),
        })
        
    df = pd.DataFrame(paises)
    
    print("\n Dataframe de paises (primeras filas):")
    print(df.head(30))
    
    
    europa_df = df[df["Region"] == "Europe"]
    europa_ordenado = europa_df.sort_values(by="Poblacion",ascending=False)
    
    print(" Dataframe de EUROPA ordenado por població")
    print(europa_ordenado.head(20))

except requests.exceptions.RequestException as e:
    print("Error al conectarse a la API ")

```
