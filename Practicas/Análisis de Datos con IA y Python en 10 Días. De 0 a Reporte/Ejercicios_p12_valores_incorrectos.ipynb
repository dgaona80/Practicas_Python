```python
import pandas as pd
import textwrap
from rich.console import Console
from rich.table import Table
from dateutil import parser

traduccion_meses = {
    "enero":"January", "febrero":"Fbruary", "marzo": "March",
    "abril":"April", "mayo":"May", "junio":"June",
    "julio":"July", "agosto": "August", "septiembre": "September",
    "noviembre":"November", "diciembre":"December"

}


df = pd.read_csv("ventas/eventos.csv",encoding = "Latin-1")
print(f'Dataframe original: \n{df}')

print("Dataframe original (formato 2):\n")
print(textwrap.indent(df.to_string(), "    "))

#print("Dataframe original (formato 3):\n")
console = Console()

#crear tabla con estilo DATAFRAME ORIGINAL
table_orig = Table(title="Dataframe formato 3 ORIGINAL", title_style="bold cyan")

table_orig.add_column("Id",style="yellow",justify="center")
table_orig.add_column("Nombre",style="yellow",justify="center")
table_orig.add_column("Edad",style="yellow",justify="center")
table_orig.add_column("Gasto",style="yellow",justify="center")
table_orig.add_column("Fecha_Registro",style="yellow",justify="center")
table_orig.add_column("SuscrupciГіn",style="yellow",justify="center")

for _, row in df.iterrows():
    table_orig.add_row(str(row["Id"]), str(row["Nombre"]), str(row["Edad"]), str(row["Gasto"]), str(row["Fecha_Registro"]), str(row["Suscripcion"]))

#Dibujar dataframe original
console.print(table_orig)


print("\n\nTipos de dato (original):\n",df.dtypes)

"""
Cambios:
Edad 
Gasto
Fecha registro
SuscripciГіn

"""
df["Edad"] = pd.to_numeric(df["Edad"],errors = "coerce")
df["Gasto"] = pd.to_numeric(df["Gasto"],errors = "coerce")

#campo de fecha datatime
#df["Fecha_Registro"] = pd.to_datetime(df["Fecha_Registro"],errors = "coerce",dayfirst=True)

def traducir_fecha(fecha):
    if pd.isna(fecha):
        return fecha
    fecha = str(fecha)

    for esp, eng in traduccion_meses.items():
        if esp in fecha.lower():
            return fecha.lower().replace(esp, eng).title()
    return fecha


df["Fecha_Registro_Traducida"] = df["Fecha_Registro"].apply(traducir_fecha)

def parsear_fecha_mixta(fecha):
    try:
        return parser.parse(str(fecha), dayfirst = False)
    except:
        return pd.NaT

df["Fecha_Registro_Parseada"] = df["Fecha_Registro_Traducida"].apply(parsear_fecha_mixta)

print("Fechas cambiadas:")
print(df["Fecha_Registro"].dtypes)
print(df["Fecha_Registro"])
print(df[["Fecha_Registro","Fecha_Registro_Traducida","Fecha_Registro_Parseada"]])
print("\nDatos cambiados (tipo)\n",df[["Fecha_Registro","Fecha_Registro_Traducida","Fecha_Registro_Parseada"]].dtypes)


#cambio de dato booleano

def convertir_booleano(valor):
    if str(valor).strip().lower() in ["true","1","si","sГӯ"]:
        return True
    elif str(valor).strip().lower() in ["false","0","no"]:
        return False

    else:
        return pd.NaT

df["Suscripcion_DQ"] =df["Suscripcion"].apply(convertir_booleano)
#crear tabla con estilo

print(df)

table = Table(title="Dataframe con proceso de calidad", title_style="bold cyan")

table.add_column("Id",style="green",justify="center")
table.add_column("Nombre",style="yellow",justify="center")
table.add_column("Edad",style="bright_green",justify="center")
table.add_column("Gasto",style="bright_blue",justify="center")
table.add_column("Fecha_Registro",style="magenta",justify="center")
table.add_column("Fecha_Registro_Parseada",style="bright_magenta",justify="center")
table.add_column("SuscripciГіn",style="white",justify="center")
table.add_column("SuscripciГіn_DQ",style="white",justify="center")

for _, row in df.iterrows():
    table.add_row(str(row["Id"]), str(row["Nombre"]), str(row["Edad"]), str(row["Gasto"]), str(row["Fecha_Registro"]), str(row["Fecha_Registro_Parseada"]) ,str(row["Suscripcion"]),str(row["Suscripcion_DQ"]))

#mostrar datos en consola
console.print(table)
```
