```python
import pandas as pd
import numpy as np

data = {
    "Cliente": ["Ana","Luis","Carla","Pedro","Laura"],
    "Edad" :[25,42,19,35,66],
    "Gasto_Total":[120,350,80,560,200],
    "Fecha_Registro":["2023-01-01","2022-11-12","2023-06-15","2021-03-20","2022-08-05"]
}

df = pd.DataFrame(data)

df["Fecha_Registro"] = pd.to_datetime(df["Fecha_Registro"])

print("-------------- Dataframe original:   ----------------------")

print(df)
#crear nueva columna np.where(condicion) 

df['Es_mayor'] = np.where(df['Edad'] >= 18,'Sí','No')

#Otra forma:

df['Es_mayor_v2'] = df['Edad'] >=18

print("\n-------------- Dataframe con nueva columna  ----------------------")

print(df)

condiciones = [
    df['Gasto_Total'] < 200,
    (df['Gasto_Total'] >=200) & (df['Gasto_Total']<=600),
    df['Gasto_Total'] > 600,
]

segmentos = ['Bajo','Medio','Alto']

df['Segmento'] = np.select(condiciones,segmentos,default ='')

print("\n-------------- Clasificación de gastos por rangos  ----------------------\n",df)

df['Mes_Registro'] = df['Fecha_Registro'].dt.month
df['Año_Registro'] = df['Fecha_Registro'].dt.year

print("-------------- Nueva columna de mes  ----------------------\n",df)
print("--------------   ----------------------")
print("--------------   ----------------------")
```
