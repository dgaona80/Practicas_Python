```python
import pandas as pd

data = {
    "Estudiante" : ["Ana","Luis","Pedro"],
    "Trim1" : [85,78,92],
    "Trim2" : [88,74,90],
    "Trim3" : [90,80,85],
    
}

df = pd.DataFrame(data)

print("------------- Dataframe original:----------\n",df)

#melt combina filas en columnas

df_largo = pd.melt(df,id_vars = ["Estudiante"], var_name="Trimestre",value_name="Calificacion")  #indice general , poner columna

print("\n------------- Dataframe largo:----------\n",df_largo)

#pivot
df_ancho = df_largo.pivot(index="Estudiante",columns="Trimestre",values="Calificacion").reset_index()  #pivote

print("\n------------- Dataframe ancho:----------\n",df_ancho)

#pivot table

df_promedio = df_largo.pivot_table(
    index = "Estudiante",
    columns="Trimestre",
    values="Calificacion",
    aggfunc="mean").reset_index()

print("\n------------- Dataframe promedio:----------\n",df_promedio)


#restructuración flexible (mas sensible

df_stack = df.set_index("Estudiante").stack().reset_index()
df_stack.columns = ["Estudiante","Trimestre","Calificacion"]


print("\n------------- Dataframe stack:----------\n",df_stack)
```
