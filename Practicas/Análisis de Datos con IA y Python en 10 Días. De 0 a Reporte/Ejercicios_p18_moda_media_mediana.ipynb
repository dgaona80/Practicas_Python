```python
import pandas as pd
import numpy as np

data = {
    "Empleado" : ["Ana","Luis","Pedro","María","Sofía","Juan","Laura","Carlos"],
    "Salario" : [1500,1600,1700,1800,1900,2000,2100,10000],   
}

df = pd.DataFrame(data)

#MEDIA sí está influenciada por los outliers

print("------------- Dataframe original:----------\n",df,"\n\n",df.describe())

media = df["Salario"].mean()

print("------------- Media de salario:----------\n",media)

#MEDIA 2825 , indica que los datos están alterados por un valor extremo


#MEDIANA valor representativo apra el analisis, incluyendo los outliers

mediana = df["Salario"].median()
print("------------- Mediana de salario:----------\n",mediana)

#MODA valor que más veces aparece en los datos

moda = df["Salario"].mode()
print("------------- Moda de salario:----------\n",moda)


```
