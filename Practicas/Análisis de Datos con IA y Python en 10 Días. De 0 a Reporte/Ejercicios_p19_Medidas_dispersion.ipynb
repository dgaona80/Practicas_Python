```python
import pandas as pd
import numpy as np

data = {
    "VENTAS_A" : [50,50,50,50,50],
    "VENTAS_B" : [10,30,50,70,90]   
}

df = pd.DataFrame(data)



print("------------- Dataframe original:----------\n",df,"\n\n",df.describe())


#Rango punto más alto y punto más bajo
rango_A =df["VENTAS_A"].max() - df["VENTAS_A"].min()
rango_B =df["VENTAS_B"].max() - df["VENTAS_B"].min()

print("------------- Rango ventas A:----------\n",rango_A)
print("------------- Rango ventas B:----------\n",rango_B)


#Varianza cuantos se alejan los datos del promedio pero al cuadrado

var_A = df["VENTAS_A"].var()
var_B = df["VENTAS_B"].var()

print("------------- Varianza ventas A:----------\n",var_A)
print("------------- Varianza ventas B:----------\n",var_B)


#Desviación estándar raiz cuadrada de la varianza

std_A = df["VENTAS_A"].std()
std_B = df["VENTAS_B"].std()
print("------------- desviación std ventas A:----------\n",std_A)
print("------------- desviación std ventas B:----------\n",std_B)


#Percentiles, identificar

p25 = np.percentile(df["VENTAS_B"],25)
p75 = np.percentile(df["VENTAS_B"],75)

iqr =p75 -p25

print("------------- Percentil 25% ventas B:----------\n",p25)
print("------------- Percentil 75% ventas B:----------\n",p75)
print("------------- IQR de Ventas B:----------\n",iqr)
```
