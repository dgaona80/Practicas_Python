```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = {
    "Región" : ["Norte","Sur","Este","Oeste"],
    "Ventas" : [350,420,280,500]   
}

data_lineas = {
    'Mes' : ['Enero','Febrero','Marzo','abril','Mayo'],
    'Ventas' : [200,250,300,280,350]
}


#Gráfico de barras
df_lineas = pd.DataFrame(data_lineas)

df = pd.DataFrame(data)

plt.figure(figsize=(7,5))

#sns.barplot(x= 'Región',y = 'Ventas', data=df, palette='viridis')
sns.barplot(x="Región", y="Ventas", data=df, hue="Región", palette="Set2", legend=False)
plt.title('Ventas por región')
plt.xlabel('Región')
plt.ylabel('Ventas')
plt.show()


#Gráfico de líneas

plt.figure(figsize=(8,6))
sns.lineplot(x='Mes', y = 'Ventas', data = df_lineas, marker='*')
plt.title('Tendencias de ventas mensuales por región')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.show()

```
