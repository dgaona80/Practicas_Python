```python
import pandas as pd
import os
from pathlib import Path

def attach_csv_x_carpeta(carpeta):

    # Definir la carpeta donde están los archivos:
    #carpeta = 'ventas'
    path = Path("C:/Users/USER/Documents/CURSO_PYTHON/Análisis de Datos con IA y Python en 10 Días. De 0 a Reporte/")


    #paso 2: buscar archivos con terminacion CSV en la carpeta ventas

    ruta_completa=path/carpeta
    print(f'path completo: {ruta_completa}\n')
    #encontrar archivos solamente CSV
    archivos_encontrados = [archivo for archivo in os.listdir(carpeta) if archivo.endswith('.csv')]
    print(f'Archivos encontrados: {archivos_encontrados}')

    carpeta_raiz = Path(r"C:\Users\USER\Documents\CURSO_PYTHON")
    archivos_encontrados_all = [f for f in carpeta_raiz.iterdir() if f.suffix == '.csv']
    print(f'Archivos CSV encontrados desde la carpeta raiz CURSOS_PYTHON: {archivos_encontrados_all}')

    #paso 3 Crear datafrme (contenedor)

    df_csv = []

    #paso 4 leer cada archivo CSV

    for archivo in archivos_encontrados:
        ruta_completa = os.path.join(carpeta,archivo) #ventas/archivoXXXcsv
        df = pd.read_csv(ruta_completa)
        mes = archivo.split('_')[1].replace('.csv','') #seccionar el nombre y dejar el mes en el nombre del archivo
        df['Fuente'] = mes #agregar este valor en un nuevo campo que se agregará en cada inserción de datos del archivo.
        df_csv.append(df) #agregar cada set de datos al data frame temporal
    
    df_final = pd.concat(df_csv, ignore_index = True) #agregar todos los dataframes
    return df_final

df_resultado = attach_csv_x_carpeta("ventas2")
print(f'Data frame final:\n')
print(df_resultado)


```
