```python
import pandas as pd


#paso 1: lectura de las hojas del archivo excel

archivo_excel = pd.read_excel('ventas2/ventas_abril.xlsx',sheet_name = None, engine = 'openpyxl') #todas las hojas

# paso 2 mostrar archivo
print(archivo_excel.keys())
print(archivo_excel)

# paso 3: guardamos los datos en un df

df_xls = []

#paso 4 recorremos las hojas con su nombre

for nombre_hoja, df in archivo_excel.items():
    df['mes'] = nombre_hoja
    df_xls.append(df)
    
df_final = pd.concat(df_xls,ignore_index=True)

print(f'hojas de excel concatenada:\n {df_final}')

    
```
