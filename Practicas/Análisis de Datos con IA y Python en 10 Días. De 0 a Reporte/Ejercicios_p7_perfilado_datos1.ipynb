```python
import pandas as pd

df = pd.read_csv('ventas/productos.csv',encoding='latin-1')

#df = pd.read_excel('ventas/productos.csv',sheet_name = None, engine = 'openpyxl') #todas las hojas

print(f'Vista general del df')
print(df.head(10))

print(f'\nInfo general:')
df.info()

print(f'Estadísticas:')
df.describe()

print(f'\nConteos de Datos Null:')
print(df.isnull().sum())


print(f'\nDatos duplicados exactos:')
print(df.duplicated().sum())

#limpieza datos duplicados
df["nombre"] = df["nombre"].str.strip()
df["categoria"] = df["categoria"].str.strip()
df["estado"] = df["estado"].str.strip()

print(f'\nduplicados después de limpieza:')
print(df.duplicated().sum())

#duplicados específicos

duplicados = df.duplicated(subset=["nombre","categoria","estado"])
print(f'\nDatos duplicados por columnas:')
print(df[duplicados])


duplicados_categoria = df.duplicated(subset=["categoria"])
print(f'\nDatos duplicados por categoría:')
print(df[duplicados_categoria])

#valores estado
print(f'\n Valores por estado:')
print(df["estado"].value_counts())


#valores únicos x estado
print(f'\n Valores ÚNICOS por estado:')
df["estado"] = df["estado"].str.lower().str.strip()
print(df["estado"].value_counts())


#valores únicos
print(f'\n Valores únicos por categoria:')
print(df["categoria"].value_counts())
```
