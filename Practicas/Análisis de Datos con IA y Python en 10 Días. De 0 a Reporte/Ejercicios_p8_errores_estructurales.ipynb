```python
import pandas as pd

df = pd.read_csv("ventas/ventas.csv", encoding="latin1")
print(df.head(100))
print("\nPerfilamiento de productos:")
print(df["Producto"].value_counts())
print("Limpieza de datos:\n")
df["Producto"] = df["Producto"].replace(
    {
        "Zapato" : "Zapatos",
        "Camisaa" : "Camisa"
    }
)

print("Resultado de limpieza de datos:")
print(df["Producto"].value_counts())


#producto $ 5,000 = error ? o producto excepcional
print("Perfilamiento de precios:")
print(df["Precio"].describe())

Q1 = df["Precio"].quantile(0.25)
Q3 = df["Precio"].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q1 + 1.5 * IQR

print(f'Los valores fuera de rango : menor a {limite_inferior} o mayor a: {limite_superior}')

#identificar los valores que están fuera de ese rango:
outliers = df[(df["Precio"] < limite_inferior) | (df["Precio"] > limite_superior)]

print(outliers)

"""
    Opciones:

    1.- Eliminamos valores que salen del rango ( no afecta a los demás datos )
    2.- Cambiamos valores (separamos el valor correcto)
    3.- Mantenemos valores (con una justificación)

"""
```
