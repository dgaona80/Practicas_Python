# Ejercicio en clase
Tipo de cambio 30 jul

1 USD:  0.9246 EUR
1 EUR:  1.0815 USD
1 USD: $18.7628 MXN
1 EUR: $20.2898 MXN

Precio del paquete: $3,165 USD

¿Cuánto me costará en MXN?
¿Cuanto me costará en EUR?

```python
# Solución

var_CostoUSD=3165
var_Tc_MXN_USD=18.7628
var_Tc_MXN_EUR=20.2898
var_Tc_EUR_USD=1.0815
var_Tc_USD_EUR=0.9246

print("------Mi viaje a EUROPA----")
print(" ")
print("Costo del viaje: 3,165 USD")
print(" ")
print(" ")
var_Precio_MXN=var_CostoUSD*var_Tc_MXN_USD
var_Precio_EUR=var_CostoUSD*var_Tc_USD_EUR
var_Precio_MXN_EUR=var_Precio_EUR*var_Tc_MXN_EUR

print(f'Precio en MXN:     ${var_Precio_MXN:,.2f}')
print(f'Precio en EUR:       {var_Precio_EUR:,.2f}' )
print(f'Precio en MXN-EUR: ${var_Precio_MXN_EUR:,.2f}')
```
