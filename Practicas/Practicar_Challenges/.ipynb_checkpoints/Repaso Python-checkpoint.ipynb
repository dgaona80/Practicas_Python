```python
import pandas as pd
import numpy as np

# Leer datos
df = pd.read_csv('orders.csv', parse_dates=['order_date'])

# Inspección rápida — siempre haz esto primero
print('Análisis del archivo ordes.csv, total de filas, columnas:')
print(df.shape)          # (filas, columnas)
print('---------tipos de cada columna --------------------------')
print(df.dtypes)         # tipos de cada columna
print('-------------------nulos por columna:-------------------')
print(df.isnull().sum()) # nulos por columna
print('duplicados:')
print(df.duplicated().sum()) # duplicados

# Limpiar nulos
print('-------------------------limpiar nulos:----------------')
df['unit_price'].fillna(df['unit_price'].median(), inplace=True)
df.dropna(subset=['customer_id'], inplace=True)  # obligatorio
print(df)

print('------------Limpiar datos --------------------------')
# Limpiar tipos
# convierte a string y quita espacios
df['order_id'] = df['order_id'].astype(str).str.strip()
# convierte a numerico y en caso de error asigna NaN
df['qty'] = pd.to_numeric(df['qty'], errors='coerce')

# Eliminar duplicados
print('---------------eliminar duplicados-------------------------')
df = df.drop_duplicates(subset=['order_id'], keep='last')

# Renombrar columnas snake_case
print('-----------------renombrar columna catagory product-------------')
print('antes del cambio:',df.columns)
df.columns = df.columns.str.lower().str.replace(' ', '_')
print(df.columns)
```
