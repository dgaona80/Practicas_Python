```python
"""
df = pd.DataFrame({'a':[1,2,None,4],'b':[10,None,30,40]})
print(df.dropna().shape)

print(df.dropna(axis=1).shape)

print(df.dropna(how='all'))
print(df.dropna(how='all').shape)
"""

import pandas as pd

df = pd.DataFrame({
    'a': [1, 2, None, 4],
    'b': [10, None, 30, 40]
})

print("DataFrame original:")
print(df, "\n")

# Caso 1: dropna() por defecto
print("df.dropna():")
print(df.dropna())
print("Shape:", df.dropna().shape, "\n")

# Caso 2: dropna(how='all')
print("df.dropna(how='all'):")
print(df.dropna(how='all'))
print("Shape:", df.dropna(how='all').shape, "\n")

# Caso 3: dropna(axis=1)
print("df.dropna(axis=1):")
print(df.dropna(axis=1))
print("Shape:", df.dropna(axis=1).shape, "\n")

# Caso 4: dropna(thresh=1)
print("df.dropna(thresh=1):")
print(df.dropna(thresh=1))
print("Shape:", df.dropna(thresh=1).shape, "\n")

# Caso 5: dropna(thresh=2)
print("df.dropna(thresh=2):")
print(df.dropna(thresh=2))
print("Shape:", df.dropna(thresh=2).shape, "\n")
```
